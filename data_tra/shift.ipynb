```python
import pulp
import pandas as pd
from pulp import LpMaximize, LpProblem, LpVariable, LpBinary ,LpMinimize,value
import random
from datetime import datetime
```

```python
df = pd.read_csv("../data/testing.csv")

df.head()
print(df.columns)
```

```python
#df['date'] = pd.to_datetime(df['date'])
staff_level = df["Level"].unique()
date = df["date"].unique()
```

```python
shift_group = df.groupby("date").agg(list).reset_index()
```

```python
from pulp import LpProblem, LpVariable, LpMinimize, lpSum

model = LpProblem("Shift_creator", LpMinimize)
pred = [
    [10]*16,
    [12]*16,
    [14]*16,
    [15]*16,
    [16]*16,
    [18]*16,
    [20]*16
]
model += lpSum(variable.values())  # 全スタッフの勤務時間数（1時間ごとの割当）

time_map = {
    "morning": list(range(9, 13)),
    "afternoon": list(range(13, 18)),
    "night": list(range(18, 25))
}

variable = {}

for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]
    
    for i, staff_id in enumerate(ids):
        for shift_name, hours in time_map.items():
            if row[shift_name][i]:  
                for hour in hours:
                    var_name = f"x_{staff_id}_{date}_{hour}"
                    variable[var_name] = LpVariable(var_name, cat='Binary')
```

```python
for shift_i, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]
    levels = row["Level"]
    pred_levels = pred[shift_i]

    for hour in range(9, 25):
        level_sum = []
        for staff_i, staff_id in enumerate(ids):
            var_name = f"x_{staff_id}_{date}_{hour}"
            if var_name in variable:
                staff_level = levels[staff_i]
                level_sum.append(staff_level * variable[var_name])

        required_level = pred_levels[hour - 9]
        model += lpSum(level_sum) >= required_level, f"level_requirement_{date}_{hour}"
```

```python
from collections import defaultdict

staff_hour_vars = defaultdict(list)

for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]
    statuses = row["status"]
    
    for i, staff_id in enumerate(ids):
        if "留学生" in statuses[i] or "International Student" in statuses[i]:
            for hour in range(9, 25):
                var_name = f"x_{staff_id}_{date}_{hour}"
                if var_name in variable:
                    staff_hour_vars[staff_id].append(variable[var_name])

for staff_id, hour_vars in staff_hour_vars.items():
    model += lpSum(hour_vars) <= 28, f"max_hours_international_{staff_id}"
```

```python
for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]
    statuses = row["status"]
    
    for i, staff_id in enumerate(ids):
        status = statuses[i]
        
        if "高校生" in status or "Highschool" or row["Age"][i] <= 18 in status:
            for hour in range(22, 25):
                var_name = f"x_{staff_id}_{date}_{hour}"
                if var_name in variable:
                    model += variable[var_name] == 0, f"highschool_ban_{staff_id}_{date}_{hour}"
```

```python
from pulp import PULP_CBC_CMD
model.solve(PULP_CBC_CMD(msg=False))
import pandas as pd

results = []

for var_name, var in variable.items():
    if var.value() == 1:
        _, staff_id, date, hour = var_name.split("_")
        results.append({
            "staff_id": staff_id,
            "date": date,
            "hour": int(hour)
        })

```

```python
for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]

    for i, staff_id in enumerate(ids):
        for hour in range(9, 23):
            h1 = f"x_{staff_id}_{date}_{hour}"
            h2 = f"x_{staff_id}_{date}_{hour + 1}"
            h3 = f"x_{staff_id}_{date}_{hour + 2}"

            if all(h in variable for h in [h1, h2, h3]):
                model += variable[h1] + variable[h2] + variable[h3] >= 3 * variable[h1], \
                         f"consecutive_{staff_id}_{date}_{hour}"
```

```python
for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]
    
    for i, staff_id in enumerate(ids):
        hour_vars = []
        
        for hour in range(9, 25):
            var_name = f"x_{staff_id}_{date}_{hour}"
            if var_name in variable:
                hour_vars.append(variable[var_name])
        
        if len(hour_vars) >= 6:
            model += lpSum(hour_vars) <= 5 + (1 - lpSum([1 if h in variable else 0 for h in [
                f"x_{staff_id}_{date}_{h}" for h in range(9, 25)
            ]])) * 100, f"break_after_6h_{staff_id}_{date}"
```

```python
for _, row in shift_group.iterrows():
    date = row["date"]
    ids = row["ID"]

    for i, staff_id in enumerate(ids):
        for start in range(9, 20):
            block_hours = list(range(start, start + 6))
            block_vars = []

            for hour in block_hours:
                var_name = f"x_{staff_id}_{date}_{hour}"
                if var_name in variable:
                    block_vars.append(variable[var_name])

            if len(block_vars) == 6:
                model += lpSum(block_vars) <= 5, f"rest_required_{staff_id}_{date}_{start}"
```
